### Árvore Binária k-D

### A arovre k-D ou árvore k-dimensional é uma estrutura de dados de partição de espaço que organiza pontos em um espaço $k$-dimensional, como uma árvore de busca binária, mas adaptada para dados com múltiplas coordenadas (como $x, y, z$). Ela é amplamente utilizada pra otimizar buscas espaciais, como encontrar o vizinho mais próximo de um vértice. O principal mecanismo da árvoce k-D é a divisão recursiva do espaço por meio de planos de divisão perpendiculares aos eixos coordenados. Em uma árvore 2D, por exemplo, o nível 0 divide o espaço usando o eixo $x$, o nível 1 usa o eixo $y$, o nível 2 volta ao $x$, e assim por diante, alternando em ciclos. Em cada passo, um ponto é escolhido (geralmente a mediana na dimensão de divisão para manter a árvore balanceada) e usado como nó da árvore. Pontos com valor menor nessa dimensão vão para a sub-árvore esquerda, e os maiores ou iguais vão para a direita. O resultado é uma hierarquia que particiona o espaço em regiões progressivamente menores.

### Exemplo de construção com k=2:

### No conjunto de pontos de pontos $P = \{(7, 2), (5, 4), (9, 6), (2, 3), (4, 7), (8, 1)\}$

### Nível 0 (Eixo $x$): ordena por $x$ e escolhemos a mediana (ponto raiz). Se a mediana em $x$ for $(7, 2)$, os pontos com $x < 7$ (como $(5, 4), (2, 3), (4, 7)$) vão pra esquerda, e os com $x \ge 7$ (como $(9, 6), (8, 1)$) vão pra direita

### Nível 1 (Eixo $y$): Na sub-árvore esquerda, ordena por $y$. A mediana em $y$ divide a próxima geração de nós e o processo se repete até que cada folha tenha um único ponto ou um pequeno grupo.

### A sacada da árvore k-D ta na busca. quando se procura o vizinho mais próximo de um ponto de consulta, o algoritmo percorre a árvore, guiado pelas divisões. Em vez de calcular a distância para todos os pontos, ele usa as partições do espaço para eliminar grandes regiões de busca. Se a menor distância encontrada até o momento for menor que a distância da região de um hiperplano, essa região inteira é "podada" da busca, garantindo que a complexidade de busca do vizinho mais próximo seja mais eficiente, geralmente $O(\log n)$ (no caso médio).
